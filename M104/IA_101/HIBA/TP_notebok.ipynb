```python
import matplotlib.pyplot as plt
```

```python
Q1=df[x].quantile(0.25)
Q3=df[x].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = [x for x in df["math score"] if x < lower or x > upper]
print("Valeurs aberrantes :", outliers)
plt.boxplot(df["math score"])
plt.show()
```
